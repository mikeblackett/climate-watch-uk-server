```python
# Standard libraries
import datetime as dt
import os

# Third-party libraries
import pandas as pd

# Local libraries
from utilities import climnorm
```

```python

OUT_DIR = os.path.join(os.getcwd(), "data/processed/period")

```

```python
standardPeriod = climnorm.standard()
historicPeriod = climnorm.historic()
periods = climnorm.all_periods()
start_times = [pd.to_datetime('1900-01-01').strftime('%Y%m')]
end_times = [pd.to_datetime('2000-12-31').strftime('%Y%m')]
names = ['1900–2000']
type = ['twentieth century']
id=['19002000']
for period in periods:
  start_times.append(period.start_time.strftime('%Y%m'))
  end_times.append(period.end_time.strftime('%Y%m'))
  names.append(period.start_time.strftime('%Y') + "–" + period.end_time.strftime('%Y'))
  id.append(period.start_time.strftime('%Y') + period.end_time.strftime('%Y'))
  if period.start_time == standardPeriod.start_time:
    type.append('standard')
  elif period.start_time == historicPeriod.start_time:
    type.append('historic')
  else:
    type.append('')
periods_df = pd.DataFrame({
  "id": id,
  "start_time_id": start_times,
  "end_time_id": end_times,
  "name": names,
  "type": type
})
periods_df
```

```python
os.makedirs(OUT_DIR, exist_ok=True)  
periods_df.to_csv(os.path.join(OUT_DIR, "period.csv"), index=False)
```
